(b) Next write out a new file named "clean_data.dat". The new file should contain the same data as the old file, but with uniform formatting and aligned columns.

```python
a = open('messy_data.dat')
n = 1
for lines in a:
    lines = lines.split(',')
    data = []
    for numbers in lines:
        numbers = numbers.replace(' ','')
        numbers = float(numbers)
        data.append(numbers)
    n += 1
    if n == 2:
        x = np.array([data])
    else:
        x = np.vstack((x,data))
        
out = open('clean_data.dat', 'w')
out.write("%s" %(x))
out.close()
a.close()

!cat clean_data.dat
```

(c) Now re-do the same task using NumPy's loadtxt and savetxt functions.

```python
a = open('messy_data.dat')
n = 1
data = []
for lines in a:
    lines = lines.split(',')
    for numbers in lines:
        numbers = numbers.replace(' ','')
        numbers = float(numbers)
        data.append(numbers)
        
a.close()

np.savetxt('clean_data2.dat', np.reshape(data,(100,5)), fmt='%6.3f', header='Data!')

!cat clean_data2.dat
```
