```python
def deriv(theta_current,t):  # return derivatives of the array theta
    g = 9.8 #m/s^2
    ell = 0.5 #m
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([ dtheta, domega ])

t = np.arange(0.,10.+0.01,0.01)

theta_app = lambda t: th0*np.cos(np.sqrt(g/ell)*t)
g = 9.8 #m/s^2
ell = 0.5 #m
omega0 = 0.
```

```python
th0 = 0
theta_init = np.array([th0, omega0])
theta = odeint(deriv,theta_init,t) 
plt.plot(t,theta[:,0],label=r'$\theta _0 = 0$')

th0 = np.pi/4
theta_init = np.array([th0, omega0])
theta = odeint(deriv,theta_init,t) 
plt.plot(t,theta[:,0],label=r'$\theta _0 = \pi / 4$')

th0 = np.pi/2
theta_init = np.array([th0, omega0])
theta = odeint(deriv,theta_init,t) 
plt.plot(t,theta[:,0],label=r'$\theta _0 = \pi / 2$')

th0 = 3*np.pi/4
theta_init = np.array([th0, omega0])
theta = odeint(deriv,theta_init,t) 
plt.plot(t,theta[:,0],label=r'$\theta _0 = 3\pi / 4$')

th0 = np.pi
theta_init = np.array([th0, omega0])
theta = odeint(deriv,theta_init,t)
plt.plot(t,theta[:,0],label=r'$\theta _0 = \pi$')


plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)
plt.ylim(-4,4)
plt.grid()
legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
