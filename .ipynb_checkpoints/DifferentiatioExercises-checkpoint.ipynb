```python
def twoPtForwardDiff(x,y):
    """Estimates the derivative of the arrays x and y 
    by taking the differentials going forward and dividing them"""
    dydx = np.zeros(y.shape,float)
    
    dydx[0:-1] = np.diff(y)/np.diff(x)
    dydx[-1] = (y[-1] - y[-2])/(x[-1]-x[-2])
    return dydx

def twoPtCenteredDiff(x,y):
    """Estimates the derivative of the arrays x and y 
    by taking the differentials between two points equidistant from the center
    and dividing them"""
    dydx = np.zeros(y.shape,float)
    dydx[1:-1] = (y[2:] - y[:-2])/(x[2:] - x[:-2]) 
    dydx[0] = (y[1] - y[0])/(x[1]-x[0])
    dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
    return dydx
```

(b) Use your functions to take the derivative of the function y=8x$^4$+3x over the range 0 <x< 10, with steps of 0.1. How do you know they work correctly? (Hint: compare to the analytical result).

```python
x = arange(0.,10.1,0.1)
y = 8*x**4+3*x

Exact = 32*x**3+3

forward = twoPtForwardDiff(x,y)
center = twoPtCenteredDiff(x,y)

plt.plot(x,forward,label='forward difference')
plt.plot(x,center,label='center difference')
plt.plot(x,Exact,label='exact difference')
plt.legend(loc='upper right')
plt.show()
```

The functions work correctly because they are fairly similar to the exact difference

(c) What is the accuracy of the results if you use a spacing of 0.01? (Plot the relative error)

```python
x = arange(0.,10.01,0.01)
y = 8*x**4+3*x
Exact = 32*x**3+3

forward = twoPtForwardDiff(x,y)
center = twoPtCenteredDiff(x,y)

ErrorForward = 100*abs((forward - Exact)/Exact)
ErrorCenter = 100*abs((center - Exact)/Exact)

plt.semilogy(x,ErrorForward,label='forward error')
plt.semilogy(x,ErrorCenter,label='center error')
plt.ylabel('percent error')
plt.legend(loc='upper right')
plt.show()
```
