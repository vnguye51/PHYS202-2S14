Your function should test whether there are unequal weights on the data points and return the correctly calculated uncertainties for whichever case is requested.

Use the features of numpy arrays rather than loops to take advantage of their superior computational speed.


```python
def WeightedLLSqFit(x,y,w):
    """Take in arrays representing (x,y) values for a set of 
    linearly varying data and an array of weights w.
    Perform a weighted linear least squares regression.  
    Return the resulting slope and intercept
    parameters of the best fit line with their uncertainties.

    If the weights are all equal to one, the uncertainties on 
    the parameters are calculated using the 
    non-weighted least squares equations."""
    #weight
    z = 1/w**2
    
    mask = w/w[0]
    mask = mask[:] == 1
    
    if all(mask) != True:
        slope = (sum(z)*sum(z*x*y) - sum(z*x)*sum(z*y))/(sum(z)*sum(z*x*x)-(sum(z*x))**2)
        intercept = (sum(z*x*x)*sum(z*y) - sum(z*x)*sum(z*x*y))/(sum(z)*sum(z*x*x)-(sum(z*x))**2)
        slerr = sqrt(sum(z)/(sum(z)*sum(z*x*x)-(sum(z*x))**2))
        interr = sqrt(sum(z*x*x))/(sum(z)*sum(z*x*x)-(sum(z*x))**2)
    else: 
        xave = x.mean()
        yave = y.mean()
        x2ave = (x*x).mean()
        xyave = (x*y).mean()
        denom = (x2ave - xave**2)
        slope = (xyave - xave*yave)/denom
        intercept = (x2ave*yave - xave*xyave)/denom
        delta = y - (slope*x + intercept)
        d2ave = (delta*delta).mean()
        n = len(y)
        slerr = sqrt((n-2)**(-1)*d2ave/denom)
        interr = sqrt((n-2)**(-1)*d2ave*x2ave/denom)

    return slope,slerr,intercept,interr
    
```

```python
#Gaussian plus offset
def model(t,B,A,mu,sigma,A1,mu1,sigma1,A2,mu2,sigma2):
    return B + A*exp( - ((t-mu)/sigma)**2) + A1*exp(-((t-mu1)/sigma1)**2) + A2*exp(-((t-mu2)/sigma2)**2)
```
