```python
def twoPtForwardDiff(x,y):
    """Estimates the derivative of the arrays x and y 
    by taking the differentials going forward and dividing them"""
    dydx = np.zeros(y.shape,float)
    
    dydx[0:-1] = np.diff(y)/np.diff(x)
    dydx[-1] = (y[-1] - y[-2])/(x[-1]-x[-2])
    return dydx

def twoPtCenteredDiff(x,y):
    """Estimates the derivative of the arrays x and y 
    by taking the differentials between two points equidistant from the center
    and dividing them"""
    dydx = np.zeros(y.shape,float)
    dydx[1:-1] = (y[2:] - y[:-2])/(x[2:] - x[:-2]) 
    dydx[0] = (y[1] - y[0])/(x[1]-x[0])
    dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
    return dydx
```

(b) Use your functions to take the derivative of the function y=8x$^4$+3x over the range 0 <x< 10, with steps of 0.1. How do you know they work correctly? (Hint: compare to the analytical result).

$$dydx[i]≃y[i−2]−8y[i−1]+8y[i+1]−y[i+2]/12h$$

where h is the spacing between points on the x axis (assumed to be equal). This comes from keeping higher order terms from the Taylor expansion used to define the derivative. This method can be used to obtain dydx[2:-2]. In order to calculate values for dydx[0], dydx[1], dydx[-1] and dydx[-2] you must use the lowest order finite difference method (like the ones you defined in your twoPtForwardDiff(x,y) and twoPtCenteredDiff(x,y) functions).
```python
def fourPtCenteredDiff(x,y,h):
    """Estimates the derivative of the arrays x and y 
    using a Taylor expansion"""
    dydx = np.zeros(y.shape,float)
    dydx[1] = (y[2] - y[-2])/(x[2] - x[-2]) 
    dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
    dydx[0] = (y[1] - y[0])/(x[1]-x[0])
    dydx[-2] = (y[-2] - y[2])/(x[-2] - x[2])
    dydx[2:-2] = (y[0:-4]-(8*(y[1:-3])) + 8*(y[3:-1]) - (y[4:]))/(12*h)
    return dydx

h = .01
x = np.arange(0,10.01,h)
y = 8*(x**4) + (3*x)

Exact = 32*(x**3) + 3
four = fourPtCenteredDiff(x,y,h)

ErrorCenter = 100*abs((four - Exact)/Exact)

plt.plot(x,ErrorCenter,label='error')
plt.axis([0,10,0,.1])
plt.ylabel('% Error')
plt.xlabel('X')
```
