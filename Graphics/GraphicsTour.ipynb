### Simple plots

```python
#create the data to be plotted
x = np.linspace(0, 2*np.pi, 300)
y = np.sin(x)
y2 = np.sin(x**2)
```

```python
#Now plot it
plt.plot(x,y)
plt.plot(x,y2)
plt.show()
```

```python
#decrease teh number of points to illustrate the use of markers
x=np.linspace(0,2*np.pi,50)
y=np.sin(x)
y2=np.cos(x)
plt.plot(x,y,'o',markersize=5,color='r');
plt.plot(x,y2,'^',markersize=5,color='b');
```

```python
#back to our original data
x = np.linspace(0,2*np.pi,300)
y = np.sin(x)
y2=np.sin(x**2)

plt.plot(x,y)
plt.plot(x,y2)

plt.grid()

plt.savefig("Example.pdf")
plt.savefig("Example.png")
```

### Refining our plots

```python
#create the data
x = np.linspace(-np.pi, np.pi, 56, endpoint=True)
c,s= np.cos(x), np.sin(x)
```

```python
plt.plot(x,c)
plt.plot(x,s)
plt.show()
```

```python
#Create a figure of size 8x6 points, 80 dots per inch
plt.figure(figsize=(8,6),dpi=80)

#Create a new subplot from a grid of 1x1
plt,subplot(1,1,1)

#Plot cosine with a blue continous line of width 1 (pixels)
plt.plot(x,c,color="blue",linewidth=1.0,linestyle="-")

#Set x limits
plt.xlim(-4.0,4.0)

#Set x ticks
plt.xticks(np.linspace(-4,4,9,endpoint=True))

#Set y limits
plt.ylim(-1.0,1.0)

#Set y ticks
plt.yticks(np.linspace(-1,1,5,endpoint=True))

#Save figure using 72 dots per inch
plt,savefig("plot_example.png",dpi=72)

#Show result on screen
plt.show()
```

### Changing color and line widths

```python
plt.figure(figsize=(10,6),dpi=80)
plt.plot(x,c,color="blue",linewidth=2.5,linestyle="-")
plt.plot(x,s,color="red", linewidth=2.5,linestyle="-")
plt.show()
```

### Setting limits

```python
plt.figure(figsize=(10,6),dpi=80)
plt.plot(x,c,color="blue",linewidth=2.5,linestyle="-")
plt.plot(x,s,color="red",linewidth=2.5,linestyle="-")
#-----Start new code
plt.xlim(x.min() *1.1,x.max()*1.1)
plt.ylim(c.min()*1.1,c.max()*1.1)
#-----End new code
plt.show()
```

### Setting tick labels

```python
plt.figure(figsize=(10,6), dpi=80)
plt.plot(x,c,color="blue",linewidth=2.5,linestyle="-")
plt.plot(x,s,color="red",linewidth=2.5,linestyle="-")
plt.xlim(x.min()*1.1,x.max()*1.1)
plt.ylim(c.min()*1.1,c.max()*1.1)

#-----Start new code
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2, np.pi],
           [r'$-\pi$', r'$0$', r'$+\pi$'])
plt.yticks([-1,0,+1],
           [r'$-1$',r'$0$',r'$+1$'])
#-----End new code

plt.show()
```
